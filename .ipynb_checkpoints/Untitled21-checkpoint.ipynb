```python
r = 0.28
rd = 0.3
I = 0.2
G_in = 10
ro = 0.3
tmax = 800
print_t = 30
L = {}
for k1 in np.arange(2, 20, 0.2):
    L[k1]={}
    for γ in np.arange(0.0, 1.01, 0.01):
        ode2_net = ODE2_network( G=G_in*k1/r, ro= ro/r, rd = rd/r, I = I, gamma=γ, k=k1, L=None, alpha=1., Q=2)
        #plt.plot(ode2_net.n(np.arange(0, int(ode2_net.L), 1)))
        #print(ode2_net.n0)
        L[k1][γ] = ode2_net.L
```

```python
I=0.2
def I_f(I, γ, k):
    return I*(1+γ*(k-1))
```

```python
I_gamma_k ={}
for k1 in np.arange(2, 20, 0.2):
    I_gamma_k[k1]={}
    for γ in np.arange(0.0, 1.01, 0.01):
        
        #plt.plot(ode2_net.n(np.arange(0, int(ode2_net.L), 1)))
        #print(ode2_net.n0)
        I_gamma_k[k1][γ] = I_f(I, γ, k1)
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dt
def model(y,t):
    if t<20/ro:
        dydt = G_in/(r*20.0) + ro*(y) -rd/r*y
    else:
        dydt = -rd/r*y
    return dydt

# initial condition
y0 = 0

# time points
t = np.linspace(0, 100, 1000)

# solve ODE
y = odeint(model, y0, t)

# plot results
plt.plot(t,y)
plt.xlabel('time')
plt.axis(ymin=0)
plt.ylabel('y(t)')
plt.show()
```

```python
k1 = 2
r=0.2
rd = 0.4
ro = 0.4
for γ in np.arange(0.0, 1.2, 0.2):
        ode2_net = ODE2_network( G=G_in*k1/r, ro= ro/r, rd = rd/r, I = I, gamma=γ, k=k1, L=None, alpha=1., Q=2)
        #plt.plot(ode2_net.n(np.arange(0, int(ode2_net.L), 1)))
        #print(ode2_net.n0)
        L = ode2_net.L
        print(L)
        def model(y,t):
            if t<L/ro:
                dydt = (1+r/ro)*(G_in/(L) + r*(y) -rd *y)
            elif t<L/ro:
                dydt = (G_in/(L)  -rd *y)
            else:
                dydt = -rd*y
            return dydt

        # initial condition
        y0 = 0

        # time points
        t = np.linspace(0, L/ro+24, 1000)

        # solve ODE
        y = odeint(model, y0, t)

        # plot results
        plt.plot(t, y, label = 'γ = '+str(format(γ, '.1f')))
        plt.xlabel('time')
        plt.axis(ymin=0)
        plt.ylabel('$n(t)$', fontsize='large')
        plt.legend(fontsize = 'x-small')
        plt.title("Q = "+ str(2))
        #plt.show()
```
