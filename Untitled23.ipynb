```python
import numpy as np
from scipy.integrate import solve_ivp

# Define parameters
r = 0.35 # Replication rate
rd = 0.4  # Natural death rate
ro = 0.3  # Obsolescence rate
I = 0.8  # Innovation rate
k = 1.0  # Constant factor
n0 = 100.0  # Initial density
delta_x = 0.1  # Spatial step size
x_i0 = 10.0  # Initial position of x_i

# Define the time span
t_span = (0, 1000)  # Start and end time

# Define the spatial domain
x_min = 0
x_max = 10
num_points = int((200 - x_min) / delta_x) + 1

# Create the spatial grid
x_grid = np.linspace(x_min, 200, num_points)

# Define the initial condition function
def initial_condition(x):
    return [n0 if 0 < i <= x_i0 else 0 for i in x]

# Define the differential equation system
def dndt(t, n):
    x_o = t * ro  # Obsolescence front position at time t
    x_i = x_i0 + r * I * k * n[-1]  # Innovation front position at time t

    # Create an array to store the derivative values
    dn_dt = np.zeros_like(n)

    # Inside the interval [x_o, x_i]
    inside_interval = (x_grid >= x_o) & (x_grid <= x_i)
    dn_dt[inside_interval] = r * np.roll(n, -1)[inside_interval] - rd * n[inside_interval]

    # Outside the interval [x_o, x_i]
    outside_interval = ~inside_interval
    dn_dt[outside_interval] = -rd * n[outside_interval]

    return dn_dt

# Set the initial condition
initial_state = initial_condition(x_grid)

# Solve the differential equations using solve_ivp
solution = solve_ivp(
    dndt,
    t_span,
    initial_state,
    t_eval=np.linspace(t_span[0], t_span[1], 10000),
    method='RK45',  # You can choose another solver method if needed
)

# Extract the solution
n_solution = solution.y

# Plot or analyze the results as needed
```
