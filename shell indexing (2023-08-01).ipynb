Numbering a graph in shells

```python
import numpy as np
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout
from itertools import chain 

def select_largest_component(adj):
    """Take the adjacency matrix of the largest connected component in the graph.
    
    Parameters
    ----------
    adj : ndarray
    
    Returns
    -------
    ndarray
    """
    G = nx.Graph(adj)
    connected_components = nx.connected_components(G)
    largest_connected_component = max(connected_components, key=len)
    largest_connected_subgraph = G.subgraph(largest_connected_component)
    return nx.adjacency_matrix(largest_connected_subgraph).toarray()

def graph_as_shells(starting_nodes, adj):
    """Group nodes into shells growing out from starting nodes.
    
    Parameters
    ----------
    starting_nodes : list
        Starting nodes for numbering.
    adj : ndarray
        Adjacency matrix.
    
    Returns
    -------
    list of lists
        Each consecutive list is a list of nodes that belong to the next shell.
    """
    assert len(starting_nodes)>0 and all([0<=n<adj.shape[0] for n in starting_nodes])
    grouped_nodes = set()
    remaining_nodes = set(range(n))

    shells = [starting_nodes]
    grouped_nodes = set(shells[0])
    while len(grouped_nodes)<n_:
        # consider all the children of the previous shell
        children = set()
        for p in shells[-1]:
            child_ix = set(np.where(adj[p])[0].tolist())
            children |= child_ix

        # remove any children that have already been grouped
        children -= grouped_nodes
        shells.append(list(children))
        grouped_nodes |= children
    return shells

def reindex_graph_by_shell(shells):
    """Give a new indexing of nodes in graph to ensure that index always increases
    shell by shell. Numbering within each shell is arbitrary.
    
    Parameters
    ----------
    shells : list of lists
        Node numbers grouped by consecutive shell.
    
    Returns
    -------
    np.ndarray
        New indices for nodes.
    """
    return list(chain.from_iterable(shells))
```

```python
n = 200

starting_nodes = [0,10]

adj = np.random.choice(2, size=(n,n), p=(.99,.01))
adj += adj.T
adj[np.diag_indices_from(adj)] = 0
adj = adj.clip(0, 1)

# remove any nodes that are not part of the connected component
adj = select_largest_component(adj)
n_ = adj.shape[0]

# remove any nodes that are not part of the connected component
adj = select_largest_component(adj)
n_ = adj.shape[0]
n_
```

```python
# break the graph in to shells and construct new directed adj matrix
shells = graph_as_shells([0], adj)
assert sum([len(s) for s in shells])==n_  # sanity check

new_ix = reindex_graph_by_shell(shells)
new_adj = adj[new_ix][:,new_ix]
new_adj[np.tril_indices_from(adj, k=-1)] = 0
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Function to remove edges that lengthen the shortest path
def remove_lengthening_edges(G, source, target):
    # Compute initial shortest path length
    initial_path_length = nx.shortest_path_length(G, source=source, target=target)

    # Track edges to be removed
    edges_to_remove = []

    # Iterate through the edges and check if removing it lengthens the shortest path
    for u, v in list(G.edges()):
        G.remove_edge(u, v)
        
        # Check if the graph remains connected and if the shortest path length has increased
        if nx.is_connected(G) and nx.shortest_path_length(G, source=source, target=target) > initial_path_length:
            edges_to_remove.append((u, v))  # Mark edge for removal
        else:
            G.add_edge(u, v)  # Restore edge if conditions are not met

    # Remove marked edges
    G.remove_edges_from(edges_to_remove)
    return G

# Generate a Watts-Strogatz small-world network
nodes = 200
k = 2  # Each node is joined with its k nearest neighbors
p = 0.1  # Rewiring probability
G = nx.watts_strogatz_graph(nodes, k, p)

# Define source and target nodes
source = 0
target = nodes // 2

# Remove edges that lengthen the shortest path
# G = remove_lengthening_edges(G, source, target)
# Number of iterations for edge removal
iterations = 3

# Iterate the removal procedure
initial_node_count = len(G.nodes())
for i in range(iterations):
    G = remove_lengthening_edges(G, source, target)
#     # Check if the graph has shrunk by a factor of 2 or more
#     if len(G.nodes()) <= initial_node_count / 2:
#         print("Stopping iteration as the graph has shrunk by a factor of 2 or more.")
#         break
```
